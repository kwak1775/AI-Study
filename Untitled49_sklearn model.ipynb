```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
data = load_iris()
```

```python
dir(data)
```

```python
iris = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(3)
```

```python
km.fit(iris)
```

```python
vars(km)
```

```python
km.labels_
```

```python
iris['target'] = km.labels_
```
