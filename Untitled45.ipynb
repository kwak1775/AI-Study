```python
age_index = titanic[titanic.age.isnull()].index
```

```python
age_titanic = titanic.drop(index=age_index)
```

```python
age_titanic
```

```python
import pandas as pd
```

```python
pd.qcut(age_titanic.age, 10)
```

```python
def age_(x):
    return int(x)//10
```

```python
age_titanic.age = age_titanic.age.map(age_)
```

```python
age_titanic.groupby(['sex','age']).sum()[['survived']]
```
