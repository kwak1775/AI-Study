```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
import pandas as pd
```

```python
a = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
b = pd.DataFrame(wine.target, columns=['target'])
```

```python
w = pd.concat([a,b],axis=1)
```

```python
ss = StandardScaler()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
w = pd.DataFrame(mms.fit_transform(w.iloc[:,:-1],w.iloc[:,-1]), columns=w.columns[:-1])
```

```python
w.boxplot()
```

```python
w.proline.map(lambda x : (x-w.proline.mean())/w.proline.std())
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier(), w.iloc[:,:-1], w.iloc[:,-1], cv=10).mean()
```

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(2)
```

```python
vars(pca)
```

```python
a= pd.DataFrame(pca.fit_transform(w.iloc[:,:-1]))
```

```python
w2 = pd.concat([a, w.target], axis=1)
```

```python
cross_val_score(KNeighborsClassifier(), w2.iloc[:,:-1], w2.iloc[:,-1], cv=10).mean()
```
